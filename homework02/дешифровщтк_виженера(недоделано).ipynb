```python
import matplotlib as plt
```

```python
def compute_correlation(letter_list, max_key_len=50):
    counts = []
    text_len = len(letter_list)
    # Проверяем длины от 1 до max_key_len
    for k in range(1, max_key_len + 1):
        count = 0
        # Сравниваем текст с самим собой, сдвинутым на k
        for i in range(text_len - k):
            if letter_list[i] == letter_list[i + k]:
                count += 1
        counts.append(count)
    return counts
```

```python
ALPH = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
N = len(ALPH)

def get_letters_only(text):
    """Оставляет только A-Z для анализа"""
    return [c.upper() for c in text if c.upper() in ALPH]

with open('cipher.txt', 'r', encoding='utf-8', errors='ignore') as f:
        cipher_text = f.read()

clean_letters = get_letters_only(cipher_text)

counts = compute_correlation(clean_letters, max_key_len=40)
plt.bar(range(25), counts[:25])

# длина 7
```
